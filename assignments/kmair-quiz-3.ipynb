United metal produces alloys B1 and B2. B1 contains 50% copper and 50% zinc. B2 contains 75% copper and 25% zinc. The net profit on B1 is \\$120/ton and on B2 is \\$100/ton. The company has 45 tons of copper available per day, and 30 tons of zinc available per day. How much of each alloy should they make to maximize their profit? What is the profit?



The problem has 2 variables, namely B1 and B2 (the amount of alloys to be made).
The optimization problem is (for all quantities in tons):

Maximize: $\$ 120 \times B1 + \$ 100 \times B2$

subject to: $0.5 \times B1 + 0.75\times B2 = Cu$  (Amount of Cu used)

$\hspace{35pt} \, 0.5 \times B1 + 0.25\times B2 = Zn$  (Amount of Zn used)

$\hspace{35pt} \, Cu <= 45, Zn <= 30$   (Maximum amount of copper and zinc)

$\hspace{35pt} \, B1 >= 45, B2 >= 0$   (Production of alloys can only be positive)

```python
from scipy.optimize import minimize
import numpy as np


def profit(B):
    B1, B2 = B          
    return -(120 * B1 + 100 * B2)

def Cu(B):
    B1, B2 = B          
    c = 0.5 * B1 + 0.75 * B2   # Cu amount utilized
    return -(c - 45)           # max Cu = 45 tons
    
def Zn(B):
    B1, B2 = B          
    z = 0.5 * B1 + 0.25 * B2   # Zn amount utilized
    return -(z - 30)           # max Zn = 30 tons

def b1(B):
    B1, B2 = B          
    return B1                  # B1 >= 0

def b2(B):
    B1, B2 = B          
    return B2                  # B2 >= 0

sol = minimize(profit, [10, 15], constraints=[{'type': 'ineq', 'fun': f} for f in [Cu, Zn, b1, b2]])

print(f'They should produce {sol.x[0]:1.3f} tons of alloy B1 and {sol.x[1]:1.3f} tons of alloy B2 daily.')
print(f'After optimizing, the maximum daily profit can be made is: $ {-profit(sol.x):1.3f}.') 
```
