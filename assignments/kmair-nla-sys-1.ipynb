There are two balls on trajectories in the *xy* plane. One ball's trajectory follows the equation:

$y = x^2 - 2x + 1$

The other ball follows a trajectory of:

$9 (x-2)^2 + 4 (y-2)^2 - 36 = 0$

Determine if it is possible for the balls to ever collide (i.e. that they have the same (x, y) coordinates at the same time). You should demonstrate graphically how many possible collisions there are, and find all of them, and plot them on your graph.

It is not obvious how to plot an ellipse, so, note that the general equation for an ellipse is:

$\frac{(x-h)^2}{a^2} + \frac{(y-k)^2}{b^2} = 1$

Then, you can plot the parametric equations:

$x = h + a cos(t)$ and $y = k + b sin(t)$

so, for a circle of radius 1, centered at the point (0.5, 2) we have:



```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

t = np.linspace(0, 2 * np.pi)

h, a = 0.5, 1
k, b = 2.0, 1
x = h + a * np.cos(t)
y = k + b * np.sin(t)
plt.plot(x, y)
plt.axis('equal')
```

```python
# Parabola

xp = np.linspace(-2, 4)
yp = xp**2 - 2 * xp + 1

# Ellipse
t = np.linspace(0, 2 * np.pi)

h, a = 2.0, 2
k, b = 2.0, 3
xe = h + a * np.cos(t)
ye = k + b * np.sin(t)
plt.plot(xp, yp, label = 'Parabola')
plt.plot(xe, ye, label = 'Ellipse')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Identifying the number of intersection points')
plt.legend()
plt.axis('equal')
plt.show()
```

It is observable that the ellipse and the parabola intersect at exactly 2 points near:
1. x = 1; and
2. x = 3

```python
from scipy.optimize import fsolve

def Dist(t):
        
    # Ellipse    
    h, a = 2.0, 2
    k, b = 2.0, 3
    x = h + a * np.cos(t)
    ye = k + b * np.sin(t)
    
    #Parabola
    yp = x**2 - 2 * x + 1
    
    Delta = yp - ye
    
    return Delta

guess = np.array([1, 3])

t_int = fsolve(Dist, guess)

print(f'The graph intersects at times: {t_int[0]:1.4f} and {t_int[1]:1.4f}.')
```

```python
#Checking for the values we get at intersection times and plotting them on the graph.

x1 = 2 + 2 * np.cos(t_int[0])
ye1 = 2 + 3 * np.sin(t_int[0])
yp1 = x1**2 - 2 * x1 + 1

print(f'At t = {t_int[0]:1.4f} we get an intersection at x = {x1:1.4f} and the value of y = {ye1:1.4f}, {yp1:1.4f}')

x2 = 2 + 2 * np.cos(t_int[1])
ye2 = 2 + 3 * np.sin(t_int[1])
yp2 = x2**2 - 2 * x2 + 1

print(f'At t = {t_int[1]:1.4f}, we get an intersection at x = {x2:1.4f} and the value of y = {ye2:1.4f}, {yp2:1.4f}.')

# Plot
```
