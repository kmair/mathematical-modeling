
## A plot



Create a plot of $f(x) = x^2$ over the range $x=0.2$.



```python
import numpy as np
x=np.linspace(0,2) #defining x in the range [0,2] for 50 points.

def f(x):
    return x**2

import matplotlib.pyplot as plt  #plotting f(x) in the given range
plt.plot(x, f(x))
plt.xlabel('x')
plt.ylabel('f(x)')
```
