When a plate is moved in a fluid at constant velocity, it causes the fluid to start moving with it. The velocity of the fluid changes with position away from the plate, and with time.

In this diagram, the plate is moving in the $x$ direction at a velocity of $U$. The velocity field at some time later is shown by the arrows in the middle.



The velocity in the $x$ direction is given by this solution to the Navier Stokes equation for this problem:

$v_x(y, t) = U[1 - erf(\frac{y}{\sqrt{4 \eta t / \rho}})]$

The error function (erf) is defined as $erf(\zeta) = \sqrt{4 / \pi} \int_0^\zeta e^{-\xi^2} d\xi$.




# Plot the velocity profile



Make a plot of $v_x$ as a function of the combined variable $z = \frac{y}{\sqrt{4 \eta t / \rho}}$ over the range of z=0 to z=2. Use `scipy.integrate.quad` to evaluate the error function.



```python
import numpy as np
from scipy.integrate import quad

def erf_int(y):
    return np.sqrt(4 / np.pi) * np.exp(-y**2)

def v_x(z):    # returns value of: Vx(z)/U 
    V = []
    for i in range(50):
        erf, err = quad(erf_int, 0, z[i])
        a = 1 - erf  # v_x 
        V.append(a)
    return V
```

```python
import matplotlib.pyplot as plt
z = np.linspace(0, 2)
plt.plot(z, v_x(z))
plt.ylabel('$v_x(z)/U$')
plt.xlabel('z')
plt.title('Velocity profile $v_x$/U as a function of combined variable z')
plt.show()
```

Provide an interpretation of what happens near $z=2$, considering that $z$ is a function of both time and distance from the plate.



The function's value and slope, both approach the value of '0' near z = 2.

Since:

<center>$\displaystyle z = \frac{y}{\sqrt{4 \eta t / \rho}}$

then at a distance of 'y' obtained on solving for different times at z = 2, we get the distance beyond which there is no effect of the plate movement and the liquid in this region doesn't move. 

Also, plotting for different values of time, we get a parabolic form of y. Thus, y basically gives us the boundary layer of fluid movement.


# Repeat the calculation using the special erf function in scipy.special.erf



Verify the solution is equivalent to the previous one.



```python
import numpy as np
from scipy.special import erf
from scipy.integrate import quad

def Vx(z):      # returns value of: Vx(z)/U
    Vx = 1 - erf(z)
    return Vx
```

```python
Vx(z)
```

```python
import matplotlib.pyplot as plt
z = np.linspace(0,2)
plt.plot(z, Vx(z))
plt.ylabel('$v_x(z)/U$')
plt.xlabel('z')
plt.title('Velocity profile $v_x$/U as a function of combined variable z')
plt.show()
```

The 2 curves can be compared together as follows:


```python
import matplotlib.pyplot as plt
plt.plot(z, v_x(z), 'o-',  label = 'Evaluating error function')
plt.plot(z, Vx(z), 'g', label = 'Using special function')
plt.ylabel('$v_x(z)/U$')
plt.xlabel('z')
plt.title('Velocity profile $v_x$/U as a function of combined variable z')
plt.legend()
plt.show()
```

The 2 curves overlap over each other and hence, both the functions give the same output.
