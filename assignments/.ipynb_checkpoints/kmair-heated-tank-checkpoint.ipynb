This tank has a heater in it and it is well-stirred. There is an inlet that allows flow into the tank, and flow out is gravity-driven.



A mass balance on the tank yields:

$A \frac{dh}{dt} = F_i - c \sqrt{h}$

The energy balance yields:

$A h \frac{dT}{dt} = F_i (T_i - T) + \frac{Q}{\rho c_p}$

The following constants are provided (you can assume the units are consistent).

|constant|value||
|---
|h||liquid height|
|A|10|tank area|
|c|1|valve constant|
|cp|1|heat capacity|
|&rho;|1|density|
|F<sub>i</sub>|1|flow in|
|Q|1|heat in|

After a long time of operation, the tank has reached a steady state height of 1, and a temperature of 30 with $Ti=29$. However, at $t=0$ there is a disturbance, and suddenly $Ti=30$. Use the information above to plot the height and temperature of the tank as a function of time in separate plots, and estimate how long it will take to reach a new steady state value. Explain why the height and temperature change the way they do.



```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

A = 10
c = 1
cp = 1
rho = 1
Fi = 1
Q = 1
T0 = 30
h0 = 1
y0 = [T0, h0]

Ti = 30   # at t = 0
yin = np.array([T0, h0])
tspan = np.linspace(0, 100)

# Solving differential equation:
def ode(y, t):
    T, h = y
    dTdt = (Fi * (Ti - T) + Q / (rho * cp)) / (A * h)
    dhdt = (Fi - c * np.sqrt(h)) / A
    return [dTdt, dhdt]

sol3 = odeint(ode, y0, tspan)

T = sol3[:, 0]
h = sol3[:, 1]

# Estimation of steady state
for i in range(50):
    diff = 31.0 - T[i]       # Since 31.0 is the steady state temperature
    if diff < 0.001:
        global t_o
        global T_o
        t_o = tspan[i]
        T_o = T[i]
        print(f'For tolerance of 0.001 temperature, at time {tspan[i]:1.3f} the temperature is {T[i]:1.3f} and can be said to have reached steady state.')
        break

# Plot of temperature
plt.plot(tspan, T.T, label = 'Temperature')
plt.plot(t_o, T_o, 'ro', label = 'Point of steady state')
plt.title('Temperature variation with time')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.legend()
plt.show()
```

When $T_i$ = 30, the value of differential equation in T becomes: $\displaystyle Ah\frac{dT}{dt} = 0 + 1$.


Thus, the temperature increases till the time the differential equation turns 0 (i.e. when T = 31).


```python
# Plot of height
import matplotlib.pyplot as plt

plt.plot(tspan, h, label='Height')
plt.xlabel('Time')
plt.ylabel('Height')
plt.legend(loc='best')
plt.title('Height as a function of time')
plt.grid()
plt.show()
```

The height remains constant with time because in the equation : $\displaystyle\frac{dh}{dt} = (F_i - c \sqrt{h})$
, we don't have any function of temperature.

While the temperature changes, in the dT/dt equation, the height remains constant as h = 1 while the Temperature changes.
