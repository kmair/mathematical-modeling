The modified Rosenbrock equation:

$f(x_1, x_2) = 100 (x_2 - x_1^2)^2 + (6.4 (x_2 - 0.5)^2 - x_1 -0.6)^2$ has three minima in the range of $-2 < x_1 < 2$ and $-2 < x_2 < 2$. Find these minima and their values.



```python
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt

def rose(X):
    x1, x2 = X
    return 100 * (x2-x1**2)**2 + (6.4* (x2 - 0.5)**2 - x1 - 0.6)**2

x1 = np.linspace(-2, 2)  
x2 = np.linspace(-2, 2)  

X1, X2 = np.meshgrid(x1, x2)  

Z = rose([X1, X2])

plt.contour(X1, X2, Z, levels=np.linspace(-60, 1000, 100))
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('Contour plot of f($x_1$, $x_2$)')
plt.colorbar()
```

```python
from autograd import grad, elementwise_grad, hessian
df = elementwise_grad(rose)
hf = hessian(rose)
```

There seems to be 3 roots at points: [0,0], [-1.0,1.0] and [1.0,1.0].

```python
G1 = np.array([0,0])
ans1 = minimize(rose, G1, jac=df, hess=hf, method='dogleg')
print(f'At point x1 = {ans1.x[0]:1.3f} and x2 = {ans1.x[1]:1.3f}, the function value is {rose(ans1.x):1.3f}.')
```

```python
G2 = np.array([-1.0,1.0])
ans2 = minimize(rose, G2, jac=df, hess=hf, method='dogleg')
ans2.x
print(f'At point x1 = {ans2.x[0]:1.3f} and x2 = {ans2.x[1]:1.3f}, the function value is {rose(ans2.x):1.3f}.')
```

```python
G3 = np.array([1.0,1.0])
ans3 = minimize(rose, G3, jac=df, hess=hf, method='dogleg')
print(f'At point x1 = {ans3.x[0]:1.3f} and x2 = {ans3.x[1]:1.3f}, the function value is {rose(ans3.x):1.3f}.')
```
