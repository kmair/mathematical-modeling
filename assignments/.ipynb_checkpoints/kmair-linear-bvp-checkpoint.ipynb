This boundary value problem is linear.

$y'' + y = 0$

With boundary conditions $y(0) = 0$ and $y(\pi/2) = 2$

First, derive the equations that approximate this equation using finite differences. You should write out the equations in a form that allow you to construct a set of linear equations, e.g. $\mathbf{A} \mathbf{x} = \mathbf{b}$.



$\displaystyle\frac{d^2 y}{dx^2} + y= 0$

Using finite differences and taking dx = h with N points, the equations can be approximated to:

$\displaystyle\frac{y_{i+1} - 2y_i + y_{i-1}}{h^2} + y_i = 0 \,\,\, \forall i$

$y_{i+1} + (h^2 - 2)y_i + y_{i-1} = 0 \,\,\, \forall i$

Thus, for i = 1 to N, we get:

$y_2 + (h^2 - 2)y_1 + y_0 = 0$

$\vdots$

$y_N + (h^2 - 2)y_{N-1} + y_{N-2} = 0$

Writing this set of linear equations as **Ay = b** where:

**y** = $[y_1, y_2, ..., y_{N-1}]$

$\mathbf{A} = \left[\begin{array}{ccccc}
 h^2-2 & 1 & 0 & ... & 0 \\
 1 & h^2-2 & 1 & ...& 0\\
 \vdots\\
 0 & ... & 0 & 1 & h^2-2
 \end{array}\right]$

**b** = $[0, 0, ..., 0]$

Next, write code to construct $\mathbf{A}$ and $\mathbf{b}$, and solve for $\mathbf{x}$.



```python
import numpy as np
import matplotlib.pyplot as plt

y0 = 0
yend = 2
x, h = np.linspace(0, np.pi / 2, 50, retstep = True)

I = np.eye(len(x)-2) * (h**2 -2)
L = np.diag(np.ones(len(x) - 3), -1) # lower diagonal
U = np.diag(np.ones(len(x) - 3), 1) # upper diagonal
A = I+L+U

b = np.zeros(len(x)-2)
b[0] -= y0
b[-1] -= yend

ans = np.linalg.solve(A,b)
y = np.concatenate([[y0], ans, [yend]])
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Solution using Finite Element method')
plt.show()
```

Show some evidence that your solution is reasonable.



```python
import numpy as np
from scipy.integrate import solve_bvp
# We can solve this by solve_bvp to verify
x = np.linspace(0, np.pi/2)

def eq(x, y):
    y, y1= y
    dydx = y1
    dy1dx = -y
    return [dydx, dy1dx]

def BC(L, U):
    yl, y1l = L
    yu, y1u = U
    return [yl, yu - 2]

# Fitting y as a line for guess

par = np.polyfit([0, np.pi/2], [0, 2], 1)
yguess = np.polyval(par, x)
y1guess = np.gradient(yguess, x)

Yg = np.array([yguess, y1guess])

ans = solve_bvp(eq, BC, x, Yg)
y_sb = ans.y[0]

plt.plot(x, y_sb)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Solution using solve_bvp')
plt.show()
```

Finally, there is a known analytical solution to this equation: $y = 2 \sin x$. Compare your answer to the known solution.



```python
y_anal = 2 * np.sin(x)
plt.plot(x, y_anal, 'ro', label = 'Analytical solution')
plt.plot(x, y_sb, 'g*', label = 'Solve_bvp')
plt.plot(x, y, 'k', label = 'Finite Element Method')
plt.legend(loc = 'best')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Comparison of both solutions to analytical solution')
plt.show()
```

Thus, since the solution from all 3 methods overlap, we can say that the solution obtained is correct.
