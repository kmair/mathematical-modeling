The following data shows the height of a genetically modified tomato plant versus the number of weeks after it was planted outdoors.



```python
import numpy as np
week = np.array([1, 2, 4, 6, 8, 10])
height = np.array([9, 15, 22, 33, 44, 52])
```

Researchers have suggested that the S-shaped logistic curve:

$height = \frac{k}{1 + e^{a + b week}}$

would be appropriate to represent this data. Use `scipy.optimize.minimize` to find the parameters $k, a, b$ in this equation. Then, use the model to plot how tall the plants will be after 24 weeks.

Discuss whether you think the S-shaped logistic curve is justified for this problem compared to a simple line, and how your prediction of the height in 24 weeks might change if you chose another model.



```python
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from scipy.optimize import curve_fit

# S-curve fit
def ht(par, w):
    k, a, b = par
    h = k / (1 + np.exp(a + b * w))
    return h

def error(par):
    e = height - ht(par, week)
    return np.sum(e**2)

p0 = [15, 4, -4]

ans = minimize(error, p0)

k, a, b = ans.x
print(f'The parameters fitted into the S-shaped curve are k = {k:1.4f}, a = {a:1.4f} and b = {b:1.4f}.')

# Simple line
def ht2(week, m, c):
    #a, b = par
    return m * week + c
    
    
m0 = (height[5]-height[0]) / (week[5] - week[0])
initial_guess = [m0, 0]

popt, pcov = curve_fit(ht2, week, height, p0=initial_guess)
mf, cf = popt
print(f'The parameters fitted into the linear curve curve are slope = {mf:1.4f} and intercept = {cf:1.4f}.')

def error2(m, c):
    e = height - ht2(week, mf, cf)
    return np.sum(e**2)


#Plot
plt.plot(week, height, 'bo')
wfit = np.linspace(min(week), max(week))
plt.plot(wfit, ht(ans.x, wfit), label = 'S-curve')
plt.plot(wfit, ht2(wfit, mf, cf), label = 'Line')
plt.xlabel('Week')
plt.ylabel('Height')
plt.title('Curve fitting for the height of plant')
plt.legend()
plt.show()
```

```python
print(f'The summed square of errors with S-curve is:{error(ans.x):1.4f} while for the line it is: {error2(mf, cf):1.4f}.')

plt.plot(week, height, 'bo')
w = np.linspace(min(week), 24)
plt.plot(w, ht(ans.x, w), label = 'S-curve')
plt.plot(w, ht2(w, mf, cf), label = 'Line')
plt.xlabel('Week')
plt.ylabel('Height')
plt.title('Curve fitting for the height of plant')
plt.legend()
plt.show()


print(f'''As seen in the plot above, after 24 weeks, the height of plant stabilizes to {ht(ans.x, w[-1]):1.3f} whereas a 
linear approximation returns height of {ht2(w[-1], mf, cf):1.3f}.''')
```

Thus, the S-curve not only gives us a lesser sum squared error, it is logically the better approximation since the height of the plant plateaus off after attaining height of almost 64 compared to a perpetually growing plant obtained by a linear plot.
